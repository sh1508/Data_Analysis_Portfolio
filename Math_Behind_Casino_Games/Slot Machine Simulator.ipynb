## How Slot Machines Work
Slot machines generally have three or more "reels," each of which has a number of symbols. While physical slot machines may have 20 or more symbols per reel, digital technology allows them to have many more — some have 256 virtual symbols — with millions of possible combinations. The combinations of symbols that pay out if you bet on them are called "paylines."

Slot machines contain random number generators (RNG) that can generate thousands of numbers per second, each of which is associated with a different combination of symbols. Whether you win or lose is determined by the RNG in the exact instant you activate each play — if it matches a payline, you win. Since each spin is independent, random and unrelated to previous or future spins, it's impossible to predict what will happen on each play.

There are may different kinds of slot machines. Some allow you to choose how many paylines to bet on per play, and how much you want to bet. Before you put your money in, figure out the cost per play, the odds, the paylines, the return to player, and anything else that will help you make the right decisions for you. Look for pay tables on or near the machine that explain everything you need to know.

## Payouts and Odds
The possible payouts and the odds of winning depend on the machine you're playing, the paylines you choose to play, and how many credits you wager. 

Machines that cost pennies to play might pay out small prizes relatively often. Others cost several dollars per play, but offer bigger jackpots and higher odds.

No matter what machine you decide to play, the odds always favour the house. This means that over time, it's more likely than not you will walk away with less money than when you started.

## House Advantage of Slot Machines
A typical average house advantage for slot machines is around 8%, meaning the average return to the player is 92%. That makes slot machines less favourable than table games, such as poker, blackjack, roulette, in terms of return to play.

## Slot Machine Simulator #1: 3 Reels

Let's assume that our slot machines have payouts for 6 combinations (lowest payout to highest): 
- 🍋🍋
- 💎💎
- 🍒🍒🍒
- 🍋🍋🍋
- 💎💎💎
- 7️⃣7️⃣7️⃣

```python
import pandas as pd
import random
import itertools
import numpy as np
import matplotlib.pyplot as plt

def spin_reel(symbols):
    return random.choice(symbols)
```

```python
slot_symbols = ["🍒", "🍒", "🍒", "🍒", "🍒", "🍒", "🍋", "🍋", "🍋", "🍋", 
                "💎", "💎", "💎", "7️⃣", "❌", "❌", "❌", "❌", "❌", "❌"]
```

```python
all_combinations_3_reels = list(itertools.product(slot_symbols, repeat=3))
num_combinations_3_reels = len(all_combinations_3_reels)

num_combinations_3_reels
```

```python
all_combinations_3_reels
```

### Probability of Each Combination

```python
count_conditions_3_reels = [
    ("🍋", 2),
    ("💎", 2),
    ("🍒", 3),
    ("🍋", 3),
    ("💎", 3),
    ("7️⃣", 3)
]

counts_3_reels = {}

for symbol, count_required in count_conditions_3_reels:
    count = sum(1 for combo in all_combinations_3_reels if combo.count(symbol) == count_required)
    counts_3_reels[(symbol, count_required)] = count

    
# Convert the counts dictionary into a DataFrame
freq_table = pd.DataFrame([
    {
        "Symbol": symbol * required,
#         "Required Count": required,
        "Frequency": count,
        "Probability": count / num_combinations_3_reels
    }
    for (symbol, required), count in counts_3_reels.items()
])

freq_table = freq_table.sort_values(by="Probability", ascending=False).reset_index(drop=True)

freq_table
```

### Step #1: Determine Payouts

Let's aim at 90% payout ratio.

```python
# Define payouts in a structured way: (symbol, count_required, payout)
payout_conditions_3_reels = [
    ("🍋", 2, 1),
    ("💎", 2, 2),
    ("🍒", 3, 10),
    ("🍋", 3, 30),
    ("💎", 3, 45),
    ("7️⃣", 3, 200)
]

total_payouts_3_reels = 0

for symbol, count_required, payout in payout_conditions_3_reels:
    count = sum(1 for combo in all_combinations_3_reels if combo.count(symbol) == count_required)
    total_payouts_3_reels += count * payout

payout_ratio_3_reels = total_payouts_3_reels / len(all_combinations_3_reels)
payouts_3_reels = { (symbol, count): payout for symbol, count, payout in payout_conditions_3_reels }

print(f"Payout Ratio: {payout_ratio_3_reels}")
```

### Step #2: Test

```python
def test_plays(num_plays, bet_choices, payouts_3_reels):
    total_bet = 0
    total_win = 0

    for _ in range(num_plays):
        bet_amount = random.choice(bet_choices)
        total_bet += bet_amount

        result = [spin_reel(slot_symbols) for _ in range(3)]

        for (symbol, count), payout_multiplier in payouts_3_reels.items():
            if result.count(symbol) == count:
                total_win += bet_amount * payout_multiplier
                break

    net_profit = total_win - total_bet
    payout_ratio = total_win / total_bet
    house_edge = 1 - payout_ratio

    print(f"Total bet: ${total_bet}")
    print(f"Total won: ${total_win:.2f}")
    print(f"Net profit: ${net_profit:.2f}")
    print(f"Payout ratio: {payout_ratio:.3f}")
    print(f"Estimated house edge: {house_edge:.1%}")

# Example usage
test_plays(num_plays=100, bet_choices=[1, 3, 5], payouts_3_reels=payouts_3_reels)
```

### Step #3: Simulate XXX Plays Per Customer for XXX Customers
#### Do we get close to our target payout ratio (0.9)?

```python
def simulate_plays(num_plays, bet_choices, payouts_3_reels):
    total_bet = 0
    total_win = 0

    for _ in range(num_plays):
        bet_amount = random.choice(bet_choices)
        total_bet += bet_amount

        result = [spin_reel(slot_symbols) for _ in range(3)]

        for (symbol, count), payout_multiplier in payouts_3_reels.items():
            if result.count(symbol) == count:
                total_win += bet_amount * payout_multiplier
                break

    payout_ratio = total_win / total_bet if total_bet > 0 else 0
    return payout_ratio
```

```python
# Simulate 100 plays per customer, for 1000 customers
num_customers = 100
num_plays_per_customer = 100
bet_choices = [1, 3, 5]

payout_ratios = []
for _ in range(num_customers):
    ratio = simulate_plays(num_plays_per_customer, bet_choices, payouts_3_reels)
    payout_ratios.append(ratio)

# Compute average payout ratio
avg_payout = np.mean(payout_ratios)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(payout_ratios, bins=30, edgecolor='black', color='skyblue')
plt.title('Distribution of Payout Ratios per Customer (100 plays each)')
plt.xlabel('Payout Ratio (Total Win / Total Bet)')
plt.ylabel('Number of Customers')

# Reference lines
plt.axvline(x=1.0, color='red', linestyle='--', label='Break Even (1.0)')
plt.axvline(x=0.9, color='green', linestyle='--', label='Target Payout (0.9)')
plt.axvline(x=avg_payout, color='blue', linestyle='--', label=f'Average Payout ({avg_payout:.3f})')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Step #4: Play!!

```python
def play_3_reels_slot(bet_amount, payouts_3_reels):
    if bet_amount not in [1, 3, 5]:
        print(f"⚠️  Invalid bet amount: ${bet_amount}. Allowed bets are $1, $3, or $5.")
        print("Slot play canceled.")
        return  # Exit early, no spin

    print(f"Your bet: ${bet_amount}")
    print("Spinning the slot machine...")

    result = [spin_reel(slot_symbols) for _ in range(3)]
    print("--------------")
    print(" | ".join(result))
    print("--------------")

    win = 0
    for (symbol, count), payout_multiplier in payouts_3_reels.items():
        if result.count(symbol) == count:
            win = int(bet_amount * payout_multiplier)
            if count == 3 and symbol == "7️⃣":
                print(f"🎉 JACKPOT! Congratulations! You won ${win}!!!")
            else:
                print(f"{count} {symbol}s! You won ${win}!")
            break

    if win == 0:
        print("Better luck next time!")
```
